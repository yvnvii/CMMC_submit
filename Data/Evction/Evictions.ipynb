```python
evictions['Neighborhood'] = evictions['UHFNAME'].map(eviction_to_residential)
```

```python
unmatched_residential = set(residential['Neighborhood']) - set(evictions['Neighborhood'])
unmatched_evictions = set(evictions['Neighborhood']) - set(residential['Neighborhood'])

print("Unmatched Residential:", unmatched_residential)
print("Unmatched Eviction:", unmatched_evictions)
```

```python
merged = residential.merge(evictions, on=['Neighborhood', 'Year'], how='inner')

# Compute Rent-to-Income Ratio (RTI) as (Median Rent / Median Income) * 100
merged["Eviction Rate"] = (merged["Evictions"]/ merged["Unit"]) * 100
```

```python
merged
```

```python
eviction_grouped = merged.groupby(['Neighborhood', 'Year'], as_index=False).agg({
    'Evictions': 'sum',
    'Unit': 'first'  # Assuming Units are constant per neighborhood-year
})
```

```python
eviction_grouped['Eviction Rate'] = eviction_grouped['Evictions'] / eviction_grouped['Unit']
```

```python
pop_file_path = "./pop.csv"

pop = pd.read_csv(pop_file_path, names = ['colnum', 'Neighborhood', 'Year', 'Population'])
pop.drop(columns = ['colnum'])
```

```python
merged_with_pop = eviction_grouped.merge(pop, on=['Neighborhood', 'Year'], how='inner')
merged_with_pop["Eviction Rate by Population"] = (merged_with_pop["Evictions"] / merged_with_pop["Population"]) * 100
merged_with_pop
```
