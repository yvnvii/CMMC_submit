```python
import pandas as pd

xls = pd.ExcelFile('./zipcode.xlsx')
modzcta_dict = xls.parse(sheet_name=0)  # adjust sheet_name if needed
modzcta_dict.to_csv('modzcta.csv', index=False)

modzcta_dict.head(), modzcta_dict.columns
```

```python
eviction['ZIPCODE'] = eviction['Zipcode'].astype(str)
modzcta_dict['ZCTA'] = modzcta_dict['ZCTA'].astype(str)

# Merge on ZIPCODE ↔ ZCTA
evictions_with_neighborhood = eviction.merge(modzcta_dict, left_on='ZIPCODE', right_on='ZCTA', how='left')
evictions_with_neighborhood
```

```python
evictions_with_neighborhood = evictions_with_neighborhood.drop(columns=["FIPS"])
```

```python
evictions_with_neighborhood = evictions_with_neighborhood.drop(columns=["ZCTA"])
evictions_with_neighborhood = evictions_with_neighborhood.drop(columns=["MODZCTA"])
evictions_with_neighborhood = evictions_with_neighborhood.drop(columns=["UHFCODE"])
```

```python
evictions_with_neighborhood
```

```python
evictions_with_neighborhood_rate = evictions_with_neighborhood.loc[evictions_with_neighborhood["Data Format"].str.contains("Rate", case=False, na=False)]
```

```python
evictions_with_neighborhood_number = evictions_with_neighborhood.loc[evictions_with_neighborhood["Data Format"].str.contains("Number", case=False, na=False)]
```

```python
residential = pd.read_csv('clean_residential_units.csv')
evictions = pd.read_csv('evictions_with_neighborhood_number.csv')

print(residential.columns)
print(evictions.columns)
```

```python
residential['Neighborhood'] = residential['Neighborhood'].str.strip().str.lower()
evictions['Neighborhood'] = evictions['UHFNAME'].str.strip().str.lower()
```

```python
unmatched = set(residential['Neighborhood']) - set(evictions['Neighborhood'])
print("Unmatched Residential Neighborhoods:", unmatched)
```

```python
from fuzzywuzzy import process
import pandas as pd

# Example - assuming residential['Neighborhood'] and evictions['Neighborhood'] cleaned
res_names = residential['Neighborhood'].unique()
eviction_names = evictions['Neighborhood'].unique()

# Loop through residential names and get best match from evictions dataset
for res_name in res_names:
    match, score = process.extractOne(res_name, eviction_names)
    print(f"Residential: {res_name} → Match: {match} (Score: {score})")
```

```python
fuzzy_map = {}
for res_name in res_names:
    match, score = process.extractOne(res_name, eviction_names)
    if score > 80:  # Threshold can be adjusted
        fuzzy_map[res_name] = match

print(fuzzy_map)
```

```python
residential['Neighborhood'] = residential['Neighborhood'].replace(fuzzy_map)
```

```python
unmatched = set(residential['Neighborhood']) - set(evictions['Neighborhood'])
print("Unmatched Residential Neighborhoods:", unmatched)
```
