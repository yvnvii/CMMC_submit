```python
comTrust = pd.read_csv("Community_Trust.csv", skiprows=6, names=["Neighborhood","Year", "Data Format", "Community Trust", "FIPS"])
empRate = pd.read_csv("Employment_Ratio.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Employment Ratio", "FIPS"])
housingCondition = pd.read_csv("Fair_to_Poor_Housing.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Fair or Poor rate", "FIPS"])
shelterPop = pd.read_csv("Families_in_Shelters.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Homeless Families", "FIPS"])
occupiedPubHousing = pd.read_csv("Occupied_Public_Housing_Rent_Regulated_Units.csv",skiprows=6, names=["Neighborhood","Housing Type","Year", "Data Format", "Number of Housing", "FIPS"])
overcrowdedRentalHousing = pd.read_csv("Overcrowded_Rental_Housing.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Overcrowsed Rental Housing", "FIPS"])
parentalEmploymentInstability = pd.read_csv("Parental_Employment_Instability.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Unstable Emp Rate", "FIPS"])
residUnitbyBuildingType = pd.read_csv("Residential_Units_by_Building_Type.csv",skiprows=6, names=["Neighborhood","Building Type","Year", "Data Format", "Data", "FIPS"])
tierIIhomelessShelterCapacity = pd.read_csv("Tier_II_Homeless_Shelter_Capacity.csv",skiprows=6, names=["Neighborhood","Year", "Data Format", "Tier II Homeless Shelter Capacity", "FIPS"])
```

```python
clean_comTrust = comTrust.drop(columns= ["Data Format", "FIPS"])
clean_comTrust.to_csv("community_trust_2015_2017.csv", index=False)
```

```python
clean_empRate = empRate.drop(columns = ["Data Format", "FIPS"])
clean_empRate.to_csv("employment_ratio.csv_2005_2022", index=False)
```

```python
clean_housingCondition = housingCondition.drop(columns = ["Data Format","FIPS"])
clean_housingCondition.to_csv("housing_condition_2011_2017.csv", index=False)
```

```python
clean_parentalEmploymentInstability = parentalEmploymentInstability.drop(columns = ["Data Format","FIPS"])
clean_parentalEmploymentInstability = clean_parentalEmploymentInstability.sort_values('Year')

clean_parentalEmploymentInstability.to_csv("parental_employment_instability_2005_2022.csv", index=False)
clean_parentalEmploymentInstability
```

```python
clean_tierIIhomelessShelterCapacity = tierIIhomelessShelterCapacity.drop(columns = ["Data Format","FIPS"])
clean_tierIIhomelessShelterCapacity.to_csv("tierII_homeless_shelter_capacity_2004_2011.csv", index = False)
```

```python
clean_shelterPop_Rate = clean_shelterPop.loc[clean_shelterPop["Data Format"].str.contains("Rate", case = False, na= False)]
clean_shelterPop_Rate = clean_shelterPop_Rate.drop(columns = ["Data Format"])
clean_shelterPop_Rate.to_csv("shelter_population.csv_2005_2015", index = False)
clean_shelterPop_Rate
```

```python
clean_occupiedPubHousing_pubHousing = clean_occupiedPubHousing.loc[clean_occupiedPubHousing["Housing Type"].str.contains("Public Housing", case = False, na = False)]
clean_occupiedPubHousing_pubHousing = clean_occupiedPubHousing_pubHousing.drop(columns = ["Data Format", "Housing Type"])
clean_occupiedPubHousing_pubHousing["Number of Public Housing"] = clean_occupiedPubHousing_pubHousing["Number of Housing"]
clean_occupiedPubHousing_pubHousing = clean_occupiedPubHousing_pubHousing.drop(columns = ["Number of Housing"])

clean_occupiedPubHousing_pubHousing.to_csv("occupiedPubHousing_2002_2017.csv", index=False)
clean_occupiedPubHousing
```

```python
clean_occupiedPubHousing_rentRegulated = clean_occupiedPubHousing.loc[clean_occupiedPubHousing["Housing Type"].str.contains("Rent Regulated", case = False, na = False)]
clean_occupiedPubHousing_rentRegulated = clean_occupiedPubHousing_rentRegulated.drop(columns = ["Data Format", "Housing Type"])
clean_occupiedPubHousing_rentRegulated["Number of Rent Regulated Housing"] = clean_occupiedPubHousing_rentRegulated["Number of Housing"]
clean_occupiedPubHousing_rentRegulated = clean_occupiedPubHousing_rentRegulated.drop(columns = ["Number of Housing"])

clean_occupiedPubHousing_rentRegulated.to_csv("occupiedRentRegulated_2002_2017.csv", index=False)
```

```python
clean_residUnitBuildingType_oneAndTwoFamily = clean_residUnitBuildingType.loc[clean_residUnitBuildingType["Building Type"].str.contains("One and Two Family", case = False, na = False)]
clean_residUnitBuildingType_oneAndTwoFamily_Percent = clean_residUnitBuildingType_oneAndTwoFamily.loc[clean_residUnitBuildingType_oneAndTwoFamily["Data Format"].str.contains("Percent", case = False, na = False)]

clean_residUnitBuildingType_oneAndTwoFamily_Percent = clean_residUnitBuildingType_oneAndTwoFamily_Percent.drop(columns = ["Data Format"])
clean_residUnitBuildingType_oneAndTwoFamily_Percent["Percentage of one and two family resident unit"] = clean_residUnitBuildingType_oneAndTwoFamily_Percent["Data"]
clean_residUnitBuildingType_oneAndTwoFamily_Percent = clean_residUnitBuildingType_oneAndTwoFamily_Percent.drop(columns = ["Data", "Building Type"])
clean_residUnitBuildingType_oneAndTwoFamily_Percent.to_csv("one_and_two_family_resident_unit_2018.csv")
```

```python
clean_residUnitBuildingType_MultiFamily_WalkUp = clean_residUnitBuildingType.loc[clean_residUnitBuildingType["Building Type"].str.contains("Multi-Family Walk-Up", case = False, na = False)]
clean_residUnitBuildingType_MultiFamily_WalkUp_Percent = clean_residUnitBuildingType_MultiFamily.loc[clean_residUnitBuildingType_MultiFamily["Data Format"].str.contains("Percent", case = False, na = False)]

clean_residUnitBuildingType_MultiFamily_WalkUp_Percent = clean_residUnitBuildingType_MultiFamily_WalkUp_Percent.drop(columns = ["Data Format"])
clean_residUnitBuildingType_MultiFamily_WalkUp_Percent["Percentage of multiple family with walk up resident unit"] = clean_residUnitBuildingType_MultiFamily_WalkUp_Percent["Data"]
clean_residUnitBuildingType_MultiFamily_WalkUp_Percent = clean_residUnitBuildingType_MultiFamily_WalkUp_Percent.drop(columns = ["Data", "Building Type"])
clean_residUnitBuildingType_MultiFamily_WalkUp_Percent.to_csv("multi_family_walkup_resident_unit_2018.csv")
```
